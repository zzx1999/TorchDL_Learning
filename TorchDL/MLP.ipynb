```python
#MLP 多层感知机
#隐藏层，输入和输出层之间，多层感知机中的隐藏层和输出层都是全连接层。
#输入X：n×d(批量大小n,输入个数d),隐藏层输出H: n×h，因为都是全连接层，设置权重Wh:d×h，偏差参数bh：1×h
#输出层权重Wo:h×q，偏差bo:1×q
#H = XWh+bh
#O = HWo+bo
#但是如果把这两个式子联立 O = XWhWo+bhWo+bo，等价一个单层神经网络，输出层权重是WhWo，偏差是bhWo+bo
#所以引入非线性变换，例如对隐藏变量使用按元素运算的非线性函数进行变换，然后再作为下一个全连接层的输入。
#这个非线性函数被称为激活函数
```

```python
%matplotlib inline
import torch
import numpy as np
import matplotlib.pylab as plt
import sys
sys.path.append("..") 
# import d2lzh_pytorch as d2l
import d2l.torch as d2l
#定义绘图函数
def xyplot(x_vals, y_vals, name):
    d2l.set_figsize(figsize=(5, 2.5))
    d2l.plt.plot(x_vals.detach().numpy(), y_vals.detach().numpy())
    #关于detach：返回一个新的tensor，从当前计算图中分离下来的，但是仍指向原变量的存放位置,不同之处只是
    #requires_grad为false，得到的这个tensor永远不需要计算其梯度，不具有grad。
    #即使之后重新将它的requires_grad置为true,它也不会具有梯度grad
    #使用detach返回的tensor和原始的tensor共同一个内存，即一个修改另一个也会跟着改变。
    d2l.plt.xlabel('x')
    d2l.plt.ylabel(name + '(x)')
```

```python
#画relu函数 max(x,0).
x = torch.arange(-8.0,8.0,0.1,requires_grad=True)
y = x.relu()
xyplot(x,y,'relu')
#relu导数函数
#当输入为负数时，ReLU函数的导数为0；当输入为正数时，ReLU函数的导数为1。
y.sum().backward()
xyplot(x,x.grad,'grad of relu')
```

```python
#画sigmoid 1/(1+exp(−x)).
y = x.sigmoid()
xyplot(x,y,'sigmoid')
#sigmoid导数:sigmoid′(x)=sigmoid(x)(1−sigmoid(x)).
x.grad.zero_()
y.sum().backward()
xyplot(x,x.grad,'grad of sigmoid')
```

```python
#tanh（双曲正切）函数可以将元素的值变换到-1和1之间
#画tanh:[1−exp(−2x)]/[1+exp(−2x)]
y = x.tanh()
xyplot(x,y,'tanh')
#画导数tanh′(x)=1−tanh^2(x).
x.grad.zero_()
y.sum().backward()
xyplot(x,x.grad,'grad of tanh')
```

```python
#所以MLP可以表示为：
#H = ϕ(XWh+bh),ϕ是激活函数
#O = HWo+bo

#MLP实现
#读取数据
batch_size = 256
train_iter,test_iter = d2l.load_data_fashion_mnist(batch_size)
```

```python
#定义模型参数
num_inputs,num_outputs,num_hiddens = 784,10,256
W1 = torch.tensor(np.random.normal(0,0.01,(num_inputs,num_hiddens)),dtype=torch.float)
b1 = torch.zeros(num_hiddens,dtype=torch.float)
W2 = torch.tensor(np.random.normal(0,0.01,(num_hiddens,num_outputs)),dtype=torch.float)
b2 = torch.zeros(num_outputs,dtype=torch.float)
params = [W1,b1,W2,b2]
for param in params:
    param.requires_grad_(requires_grad=True)
#定义激活函数，这里之间调relu函数了
#定义模型
def net(X):
    X = X.view((-1,num_inputs))
    H = (torch.matmul(X,W1) + b1).relu()
    O = torch.matmul(H,W2) + b2
    return O
#定义损失函数用softmax+交叉熵
loss = torch.nn.CrossEntropyLoss()
```

```python
#模型训练
num_epochs = 5
lr = 100.0
optimizer = None
for epoch in range(num_epochs):
    train_l_sum,train_acc_sum,n = 0.0,0.0,0
    print("begin train")
    for X,y in train_iter:
        output = net(X)
        l = loss(output,y).sum()
        #梯度下降
        l.backward()
        d2l.sgd(params,lr,batch_size)
        train_l_sum += l.item()
        
        train_acc_sum += d2l.evaluate_accuracy(net,train_iter)
        n += y.shape[0]
        
        #梯度清零
        if optimizer is not None:
            optimizer.zero_grad()
        else:
            for param in params:
                param.grad.data.zero_()
    test_acc = d2l.evaluate_accuracy_gpu(net,test_iter,device)
    print('epoch %d, loss %.4f, train acc %.3f, test acc %.3f'
              % (epoch + 1, train_l_sum / n, train_acc_sum / n, test_acc))
    
            
        
```

```python
#多层感知机简简洁实现
import torch.nn as nn
from torch.nn import init

#x形状转换功能
class FlattenLayer(nn.Module):
    def __init__(self):
        super(FlattenLayer, self).__init__()
    def forward(self, x): # x shape: (batch, *, *, ...)
        return x.view(x.shape[0], -1)
    
num_inputs,num_outputs,num_hiddens = 784,10,256
net = nn.Sequential(
    FlattenLayer(),
    nn.Linear(num_inputs,num_hiddens),
    nn.ReLU(),
    nn.Linear(num_hiddens,num_outputs)
    
    
)
for params in net.parameters():
    init.normal_(params,mean=0,std=0.01)
batch_size = 256
loss = torch.nn.CrossEntropyLoss()

optimizer = torch.optim.SGD(net.parameters(),lr=0.5)
num_epochs = 5
d2l.train_ch3(net,train_iter,test_iter,loss,num_epochs,optimizer)
```
