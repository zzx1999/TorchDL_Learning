```python
# 如果没有安装pandas，则反注释下面一行
# !pip install pandas

%matplotlib inline
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import d2l.torch as d2l

print(torch.__version__)
torch.set_default_tensor_type(torch.FloatTensor)
```

```python
train_data = pd.read_csv('./data/kaggle_house/train.csv')
test_data = pd.read_csv('./data/kaggle_house/test.csv')
print(train_data.shape)
print(test_data.shape)
#前4个样本的前4个特征、后2个特征和标签（SalePrice）
train_data.iloc[0:4, [0, 1, 2, 3, -3, -2, -1]]
```

```python
#将所有的训练数据和测试数据的79个特征按样本连结。第0列是id
all_features = pd.concat((train_data.iloc[:,1:-1],test_data.iloc[:,1:]))
#数据预处理
#归一化，让每一列数值均值为0方差为1，dtype!=object
#这里是找出哪些是数值获取其列号
numeric_features = all_features.dtypes[all_features.dtypes != 'object'].index
print(numeric_features)
#对所有数值特征减掉均值厨艺方差
all_features[numeric_features] = all_features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std()))
# 标准化后，每个数值特征的均值变为0，所以可以直接用0来替换缺失值
all_features[numeric_features] = all_features[numeric_features].fillna(0)
```

```python
#处理文本，OneHot coding
#如果一列里面有五个不同值就会创建五个不同feature，dummy_na:假设是missing data会标记
all_features = pd.get_dummies(all_features,dummy_na=True)
print(all_features.shape)
```

```python
#从pandas中提取Numpy包装成张量
n_train = train_data.shape[0]
train_features = torch.tensor(all_features[:n_train].values,dtype=torch.float32)
test_features = torch.tensor(all_features[n_train:].values,dtype=torch.float32)
train_labels = torch.tensor(train_data.SalePrice.values.reshape(-1,1),dtype=torch.float32)
```

```python
loss = nn.MSELoss()
in_features = train_features.shape[1]
#定义简单的一层线性回归
def get_net():
    net = nn.Sequential(nn.Linear(in_features,1))
    return net
#如果只估计误差y'-y那么100w的房子相比于10w的房子影响更大
#使用相对误差（y'-y）/y',有除法所以取log
def log_rmse(net,featrues,labels):
    clipped_preds = torch.clamp(net(featrues),1,float('inf')) #数据限制在infinity-1
    rmse = torch.sqrt(loss(torch.log(clipped_preds),torch.log(labels)))
    return rmse
    
```

```python
#训练 借助Adam优化器
def train(net,train_features,train_labels,test_features,test_lables,num_epochs,learning_rate,weight_decay,batch_size):
    train_ls,test_ls = [],[]
    train_iter = d2l.load_array((train_features,train_labels),batch_size)
    optimizer = torch.optim.Adam(net.parameters(),lr=learning_rate,weight_decay=weight_decay)
    for epoch in range(num_epochs):
        for X,y in train_iter:
            optimizer.zero_grad()
            l = loss(net(X),y)
            l.backward()
            optimizer.step()
        train_ls.append(log_rmse(net,train_features,train_labels))
        if test_lables is not None:
            test_ls.append(log_rmse(net,test_features,test_lables))
    return train_ls,test_ls
#K折交叉验证
def get_k_fold_data(k,i,X,y):
    assert k>1
    fold_size = X.shape[0]//k # //是整除
    x_train,y_train = None,None
    for j in range(k):
        idx = slice(j * fold_size,(j+1)*fold_size)
        x_part,y_part = X[idx,:],y[idx]#取出对应部分
        if j==i:
            x_vaild,y_valid = x_part,y_part
        elif x_train is None:
            x_train,y_train = x_part,y_part
        else:
            x_train = torch.cat([x_train,x_part],0)
            y_train = torch.cat([y_train,y_part],0)
    return x_train,y_train,x_vaild,y_valid
def k_fold(k,x_train,y_train,num_epochs,learning_rate,weight_decay,batch_size):
    train_ls_sum,valid_ls_sum = 0,0
    for i in range(k):
        data = get_k_fold_data(k,i,x_train,y_train)
        net = get_net()
        train_ls,valid_ls = train(net,*data,num_epochs,learning_rate,weight_decay,batch_size)
        train_ls_sum += train_ls[-1]
        valid_ls_sum += valid_ls[-1]
        if i==0:
            d2l.plot(list(range(1,num_epochs+1)),[train_ls,valid_ls],xlabel='epoch',ylabel='rmse',xlim=[1,num_epochs+1]
                    ,legend=['train','valid'],yscale='log')
        print(f'flod {i+1},train log rmse {float(train_ls[-1]):f},' f' valid log rmse{float(valid_ls[-1]):f}')
    return train_ls_sum/k,valid_ls_sum/k
k,num_epochs,lr,weight_decay,batch_size = 5,100,5,0,64
train_l,valid_l = k_fold(k,train_features,train_labels,num_epochs,lr,weight_decay,batch_size)
print(f'{k}-折验证：平均训练log rmse:{float(train_l):f},' f'平均验证log rmse:{float(valid_l):f}')
#正真care的是平均验证mse,通过不断调参提高准确度——k,num_epochs,lr,weight_decay,batch_size
```

```python
def train_and_pred(train_features,train_labels,test_features,test_data,num_epochs,learning_rate,weight_decay,batch_size):
    net = get_net()
    #再在完整的数据集上训练一遍
    train_ls,_ = train(net,train_features,train_labels,None,None,num_epochs,learning_rate,weight_decay,batch_size)
    d2l.plot(np.arange(1,num_epochs+1),[train_ls],xlabel='epoch',ylabel='log rmse',xlim=[1,num_epochs],yscale='log')
    print(f'train log rmse {float(train_ls[-1]):f}')
    preds = net(test_features).detach().numpy()
    test_data['SalePrice'] = pd.Series(preds.reshape(1,-1)[0])
    submission = pd.concat([test_data['Id'],test_data['SalePrice']],axis=1)
    submission.to_csv('submission.csv',index=False)
train_and_pred(train_features,train_labels,test_features,test_data,num_epochs,lr,weight_decay,batch_size)
```
